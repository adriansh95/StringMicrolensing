```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.transforms as transforms
from astropy import units as u
```

```python
# Define relevant quantities
galaxy_angular_radius = 1 * u.arcsec
# string_length = 10 * u.parsec
c = 3e8 * u.meter / u.second
c = c.to(u.parsec / u.second)
string_internal_velocity = c / np.sqrt(2)
string_obs_distance = np.logspace(-1, 6, num=17) * u.parsec
string_internal_proper_motion = string_internal_velocity / string_obs_distance * u.radian
string_internal_proper_motion = string_internal_proper_motion.to(u.arcsec / u.second)
t_cross = 2 * galaxy_angular_radius / string_internal_proper_motion
```

```python
sec_per_hour = 60**2
sec_per_day = 24 * sec_per_hour
sec_per_week = 7 * sec_per_day
sec_per_month = 30 * sec_per_day
sec_per_year = 365 * sec_per_day
sec_per_5years = 5 * sec_per_year
notable_times = [sec_per_hour, sec_per_day, sec_per_week, sec_per_month, sec_per_year, sec_per_5years]
time_labels = ["1 h", "1 d", "1 w", "1 m", "1 y", "5 y"]

fig, ax = plt.subplots(figsize=(10, 10))
ax.loglog(string_obs_distance, t_cross)
ax.set_xlabel("String - Observer Distance (Parsecs)", fontsize=18)
ax.set_ylabel("Crossing Time (seconds)", fontsize=18)
ax.tick_params(labelsize=14)
blended_transform = ax.get_yaxis_transform()
axis_transform = ax.transAxes
axis_transform_inverse = axis_transform.inverted()
ax.set_title("Time to Cross a 2 Arcsec Diameter Galaxy", fontsize=18)

for time, label in zip(notable_times, time_labels):
    ax.axhline(time, ls="-.", c='k')
    (x, y) = axis_transform_inverse.transform(blended_transform.transform((0.015, time)))
    ax.text(x, y+0.015, label, fontsize=14, transform=axis_transform)
plt.show(fig)
```
