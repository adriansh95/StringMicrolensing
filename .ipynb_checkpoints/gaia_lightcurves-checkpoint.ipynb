```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from dl import authClient as ac, queryClient as qc
from astropy.coordinates import SkyCoord

lmc_center = SkyCoord("05h23m34s", "âˆ’69d45.4m")
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
fig_dir = "/dlusers/adriansh/work/analyses/plots/strings/"
```

```python
ra_query_bounds = (lmc_center.ra.deg - 10, lmc_center.ra.deg + 10)
dec_query_bounds = (lmc_center.dec.deg - 10, lmc_center.dec.deg + 10)
lmc_ra = lmc_center.ra.deg
lmc_dec = lmc_center.dec.deg
search_radius = .2
query = f"""
SELECT source_id,band,time,mag,flux,flux_error FROM gaia_dr2.light_curves
WHERE source_id IN (SELECT source_id FROM gaia_dr2.gaia_source
WHERE 't' = Q3C_RADIAL_QUERY(ra, dec, {lmc_ra}, {lmc_dec}, {search_radius}))
"""
%time lightcurve_df = qc.query(sql=query, fmt="pandas")
```

```python
source_ids = lightcurve_df["source_id"].unique()
bands = lightcurve_df["band"].unique()
```

```python
for sid in source_ids[7:8]:
    fig, ax = plt.subplots(figsize=(12, 9))
    mask_id = lightcurve_df["source_id"] == sid
    
    
    for band in bands:
        mask_band = lightcurve_df["band"] == band # Can move this outside of loop
        mask = mask_id & mask_band

        lightcurve = lightcurve_df.loc[mask]

        ax.errorbar(lightcurve["time"], lightcurve['flux'], yerr=lightcurve["flux_error"],
                    marker='.', ms=8, label=band, capsize=5)

        ax.set_xlabel('Time', fontsize=18)
        ax.set_ylabel('Flux (e-/s)', fontsize=18)
        ax.legend(fontsize=18)
        ax.tick_params(labelsize=15)
        ax.set_title(f"Gaia Source {sid}", fontsize=22)
        ax.grid(visible=True)
#         fig.savefig(f"{fig_dir}gaia_{sid}_lightcurve.png")
    plt.show(fig)
    plt.close(fig)
```
